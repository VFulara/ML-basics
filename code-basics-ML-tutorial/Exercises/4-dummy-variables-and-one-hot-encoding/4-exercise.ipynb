```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#read the data
df = pd.read_csv('carprices.csv')
df
```

```python
#initiate one hot encoding: create dummy columns for car model
dummy = pd.get_dummies(df['Car Model'])
dummy
```

```python
# concatenate dummy frame with one hot encoded values to data read from file
merged = pd.concat([df, dummy], axis='columns')
merged
```

```python
#drop Car Model and one of dummy variable column to avoid 'dummy variable trap'
# although sklearn's linear_model is aware of dummy variable trap and it will drop it automatically, but it is a good practice do it by self
final = merged.drop(['Car Model', 'Audi A5'], axis='columns')
final
```

```python
model = linear_model.LinearRegression()
X = final.drop(['Sell Price($)'], axis='columns')
y = final['Sell Price($)']
model.fit(X.values,y.values)
```

```python
model.predict([[69000,6,1,0]])
```

```python
#To find out the score of model
model.score(X.values,y.values)
```
